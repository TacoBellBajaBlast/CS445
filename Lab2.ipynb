# Visualization 1: Count Plot

```python
plt.figure(figsize=(12,6))

# Plot count plot
sns.countplot(data=diabetes, x='Age', hue='Diabetes_012', palette='viridis')

# Label x and y
plt.xlabel('Age Group (Values 1-13, 1 = 18-24, 9 = 60-64, 13 = 80 or older)')
plt.ylabel('Count')
           
# Title and Legend
plt.title('Diabetes Distribution by Age Group')
plt.legend(title='Diabetes Status', labels=['No Diabetes', 'Prediabetes', 'Diabetes'])

# Save Visualization
plt.savefig("Jamie_DiabetesDistributionByAge.png")
plt.show()
```

## Purpose and Ingsight 
This count plot displays the distribution of diabetes across different age groups. From the visualization we can see that diabetes becomes more prevalent in older age groups. We can also see that individuals who are between 19 to 44 years of age are predominately not diagnosed with diabetes. The data also suggests that prediabetes begings to rise around the ages of 45-59. The color choice in the visualization helps to separate the three diabetes statuses and the hue makes it easy to compare the prevalence across the age groups.

# Visualization 2: Violin Plot

```python
plt.figure(figsize=(10,6))

# Plot violin plot
sns.violinplot(data=diabetes, x='Diabetes_012', y='BMI', palette='coolwarm', hue='Diabetes_012')

# Label x and y
plt.xlabel('Diabetes Status (0=No, 1=Prediabetes, 2=Diabetes)')
plt.ylabel('BMI')

# Title
plt.title('BMI Distribution by Diabetes Status')

# Save Visualization
plt.savefig("Jamie_BMIDistributionByDiabetesStatus.png")
plt.show()
```

## Purpose and Ingsight 
This violin plot displays the relationship between Body Mass Index (BMI) and diabetes status. From the visualization we can see that individuals with diabetes tend to have a higher BMI and that individuals without diabetes tend to have a lower BMI. We can see that the diabetes group has a broader violin shape, this indicates that individuals with diabetes have varying BMI values. The use of the violin shape helps represent the density of common BMI values and that use of color for the visualization, starting from blue then transitioning to red, displays an urgency or high risk. It should also be stated that the innper boxplot, the white line inside the black box, highlights the medium and interquartile range within the data.

# Visualization 3: Heatmap

```python
# Select numerical columns for correlation
corr_cols = ['HighBP', 'HighChol', 'BMI', 'HeartDiseaseorAttack', 'PhysHlth', 'MentHlth']
plt.figure(figsize=(10,8))

# Plot heatmap
sns.heatmap(diabetes[corr_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')

# Title
plt.title('Correlation Between Health Indicators')

# Save Visualization
plt.savefig("Jamie_CorrelationBetweenHealthIndicators.png")
plt.show()
```

## Purpose and Ingsight 
This heatmap displays the relationship between six different health metrics. From the visualization we can see that HighBP and HighChol have the highest postivie corrolation, which confirms the already known connection between hypertension and hyperlipdiemia in individuals with diabetes. We can also see that HeartDiseaseorAttack does have correlation with HighBP but low correlation with HighChol which could indicate that high blood is a better guage of heart attacks than high cholesterol. The use of the "coolwarm" palette helps represent negative to positive correlations.

# Visualization 4: Boxplot

```python
plt.figure(figsize=(10,6))

# Plot boxplot
sns.boxplot(data=diabetes, x='Diabetes_012', y='PhysActivity', palette='Set1')

# Label x and y
plt.xlabel('Diabetes Status (0=No, 1=Prediabetes, 2=Diabetes)')
plt.ylabel('Physical Activity (1=Yes, 0=No)')

# Title
plt.title('Physical Activity Levels by Diabetes Status')

# Save Visualization
plt.savefig("Jamie_PhysicalActivityLevelsByDiabetesStatus.png")
plt.show()
```

## Purpose and Ingsight 
This boxplot displays the relationship between physical activity levels and diabetes progression. The goal of this visualization is to see how exercise habits differ across the different diabetes statuses and if physical indactivity is a consequence of diabetes or a risk factor. From the visualization we can see that physical activity shows a decline from no diabetes to prediabetes to diabetes. As expected from the data, non-diabetic individuals have the highest activity levles and diabetic individuals have the lowest activity levels. The use of the "set1" palette helps distinguish the diabetes groups and the use of red displays the danger and risk for the diabetic group. 

# Visualization 5: Stacked Bar

```python
# Create a cross-tabulation
ct = pd.crosstab(diabetes['Diabetes_012'], diabetes['HighBP'])

# Plot stacked bar
ct.plot(kind='bar', stacked=True, figsize=(10,6), color=['lightblue', 'salmon'])

plt.xlabel('Diabetes Status (0=No, 1=Prediabetes, 2=Diabetes)')
plt.ylabel('Count')

# Title and Legend
plt.title('High Blood Pressure Prevalence by Diabetes Status')
plt.legend(title='High BP', labels=['No', 'Yes'])

# Save Visualization
plt.savefig("Jamie_HighBloodPressureByDiabetesStatus.png")
plt.show()
```

## Purpose and Ingsight 
This stacked bar chart displays the relationship between diabetes status and hypertension prevalence. The goal of this visualization is to see whether diabetes progression correlates with increased hypertension risk and how high blood pressure comorbidity rates change across the different groups. From the visualization we can see that almost 50% of non-diabetics have hypertension and that nearly 75% of diabetics have hypertension. The use of the "lightblue" and "salmon" color helps separate the those with high and low blood pressure.
